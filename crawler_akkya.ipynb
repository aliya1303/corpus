```python
import requests 
import re 
from bs4 import BeautifulSoup 
from tqdm import tqdm 
import time
import random
import webbrowser
import datetime 
 
```

```python
hreffile=open(r"links_Akkya.txt", "w", encoding="utf-8")
```

```python
corpus=open(r"‪corpus_Akkya.txt", "w", encoding="utf-8")
num = 1000
while True:
    link = "http://web-corpora.net/KazakhCorpus/search/results.php?page="+str(num)+"&sid=41327&search_language=kazakh&interface_language=en"
    try:
        r = requests.get(link) 
        table = re.split('</div>',re.split(r'<div id="contexts_div" style="overflow: auto;padding-right:5px;overflow-x:hidden;z-index:0;">', r.text)[1])[0]
        table = re.split(r'<td now="" rap="nowrap">\n<b>', table)
        for item in table[1:]:
            item1 = re.split(';</nobr>',item)[0]
            item1 = re.split(';&nbsp',item1)
            meta = [x for x in item1 if x]
            header = re.split('&nbsp', meta[0])[0]
            if header == '-':
                header = ''
            corpus.write("<source>"+header+"<source>"+"\n")
            author = re.split('&nbsp',re.split(';</b>',meta[1])[1])[0]
            if author == '-':
                author = ''
            corpus.write("<author>"+author+"</author>"+"\n")
            date = re.split('&nbsp',re.split('<nobr>',meta[2])[1])[0]
            if date == '-':
                date = ''
            corpus.write("<date>"+date+"</date>"+"\n")
            corpus.write("<textrubric>"+''+"</textrubric>"+"\n")
            corpus.write("<time>"+''+"</time>"+"\n")
            corpus.write("<textname>"+''+"</textname>"+"\n")
            corpus.write("<textid>" +''+"</textid>" +"\n")
            item2 = re.split(';</nobr>',item)[1]
            sentence = re.split('</td></tr>', re.split(r'<td style="padding: 3px 10px;">', item2)[1])[0]
            sentence = re.split('<span onmouseover=', sentence)
            soylem = ''
            for i in sentence:
                if i != '\n':
                    try:
                        i = re.split('">',i)[1]
                        soz = re.split('</span>',i)[0]
                        i = i.replace('</span>','').replace('<br>','') 
                        soylem = soylem + i + ' '
                    except:
                        pass
            corpus.write("<text>"+"\n")
            corpus.write(soylem+"\n")
            corpus.write("</text>"+"\n")
            for i in sentence:
                if i != '\n':
                    try:
                        soz = re.split('">',i)[1]
                        soz = re.split('</span>',soz)[0]
                        lemma = re.split('\]', re.split(',\[',i)[1])[0]
                        postag = re.split('\]', re.split('\],\[',i)[1])[0]
                        corpus.write("<soz>" +soz+"</soz>" +"\t")
                        corpus.write("<lemma>" +lemma+"</lemma>" +"\t")
                        corpus.write("<postag>" +postag+"</postag>" +"\n")
                    except:
                        pass
        print(link)
        hreffile.write(link+"\n")
    except:
        pass
    num-=1
```
